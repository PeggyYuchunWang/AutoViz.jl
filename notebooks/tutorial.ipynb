## ArrowCar
All vehicles can be represented using the type ArrowCar defined as such:
```
@with_kw struct ArrowCar{A<:AbstractArray{Float64}, C<:Colorant} <: Renderable
    pos::A         = SVector(0.0, 0.0)
    angle::Float64 = 0.0
    length::Float64 = 4.8
    width::Float64 = 1.8
    color::C       = COLOR_CAR_OTHER
    text::String   = "" # some debugging text to print by the car
    id::Int        = 0
end
```

````julia
car1 = ArrowCar(10.0,-DEFAULT_LANE_WIDTH,0.0, id = 1, text = "vel = 30.0", color = COLOR_CAR_EGO)
car2 = ArrowCar(40.0,0.0,0.0, id = 2, text = "vel = 25.0", color = COLOR_CAR_OTHER)
car3 = ArrowCar(70.0,-DEFAULT_LANE_WIDTH,0.0, id = 3, text = "vel = 32.0", color = COLOR_CAR_OTHER)
push!(scene, car1)
push!(scene, car2)
push!(scene, car3);
````

````julia
render(scene)
````

````julia
car4_pos = [100.0, 0.0, 0.0]
push!(scene, car4_pos)
render(scene)
````
